```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# I loaded the path_of_movies_file into a variable called path. Then I created a dataframe that holds the data of movies stored in that file 
path =r"C:\Users\HP\Desktop\imdb_top_1000.csv"
data = pd.read_csv(path)
```

```python
# I am Cleaning and preprocessing the data by removing unnecessary columns and handling missing values(using drop and dropnathe function)
cols_to_remove = ['Runtime', 'Overview']  # I am replacing these two columns because we won't use them in the analysis
data_cleaned_and_processed = data.drop(cols_to_remove, axis=1)

#I used dropna method to handle the missing values
data_cleaned_and_processed = data_cleaned_and_processed.dropna()
```

```python
# In order to explore the data of movies dataset I used a summary method called describe
data_cleaned_and_processed.describe()
```

```python
# I am listing a series of genres and the sum of each of them 
genre_counts = data['Genre'].value_counts()
print(genre_counts)
# in order to find the genre with the highest popularity, I used the idmax method with the genres's series
most_popular_genre = genre_counts.idxmax()
print(most_popular_genre)
```

```python
#Here I am creating a new data frame that contains the count of movies/TV shows in each genre using panda
Dframe_of_genre_counts = pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})
print(Dframe_of_genre_counts)
```

```python
# To Visualize the data I used a bar chart to show the most popular genre of movies/TV shows using pandas and the matplotlib library.
plt.bar(Dframe_of_genre_counts['Genre'], Dframe_of_genre_counts['Count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('The most popular genre of movies and TV shows')
plt.xticks(rotation=90)
plt.show()
```

```python
# To identify the director with the most top-rated movies/TV shows by creating a new dataframe containing the count of movies/TV shows directed by each director using pandas

series_of_director_counts = data['Director'].value_counts()

the_director_having_the_most_popularity = series_of_director_counts.idxmax()
```

```python
# To visualize the data using a bar chart to show the director with the most top-rated movies/TV shows using pandas and matplotlib.
df_of_director_counts = pd.DataFrame({'Director': series_of_director_counts.index, 'Count': series_of_director_counts.values})

# Plot the bar chart
plt.bar(df_of_director_counts['Director'], df_of_director_counts['Count'])
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('The director with the most top-rated movies/TV shows')
plt.xticks(rotation=90)
plt.show()
```
